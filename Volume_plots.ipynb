```python
import numpy as np
import matplotlib.pyplot as plt

plt.rc('text', usetex=True)
plt.rc('font', family='serif')

def E_c(L_inv):
    return 2*np.pi*L_inv
def E2_c(L_inv):
    return 4*np.pi*L_inv

def E_l(L_inv):
    return np.arccosh(2 -np.cos( 2*np.pi*L_inv) )
def E2_l(L_inv):
    return np.arccosh(2 -np.cos( 4*np.pi*L_inv) )

def E_m(L_inv,mass):
    return np.arccosh(np.cosh(mass) + 1 -np.cos( 2*np.pi*L_inv) )



x_vol=[1/8,1/16,1/24,1/32]

#en_gr_0_55=[0.747,0.367,0.264,0.2330]
#err_en_gr_0_55=[0.038,0.065,0.072,0.097]
en_gr_0_55=[0.747,0.368397,0.290,0.196]
err_en_gr_0_55=[0.052,0.068,0.029,0.018]

#en_gr_0_65=[0.7304,0.390,0.264,0.195]
#err_en_gr_0_65=[0.021,0.072,0.027,0.11]
en_gr_0_65=[0.765,0.397,0.287,0.192]
err_en_gr_0_65=[0.026,0.029,0.030,0.013]

#en_ex_0_55=[np.nan,0.609,0.566,np.nan]
#err_en_ex_0_55=[np.nan,0.084,0.069,np.nan]
en_ex_0_55=[0.967,0.623,0.557,np.nan]
err_en_ex_0_55=[0.23,0.028,0.024,np.nan]

#en_ex_0_65=[1.10,1.10,np.nan,np.nan]
#err_en_ex_0_65=[0.16,0.30,np.nan,np.nan]
en_ex_0_65=[1.102,np.nan,np.nan,np.nan]
err_en_ex_0_65=[0.098,np.nan,np.nan,np.nan]

#en_2p_0_55=[1.44,0.771,np.nan,0.4168]
#err_en_2p_0_55=[0.21,0.047,np.nan,0.074]
en_2p_0_55=[1.50,0.754,np.nan,0.371]
err_en_2p_0_55=[0.22,0.091,np.nan,0.048]

#en_2p_0_65=[np.nan,0.772,0.528,0.427]
#err_en_2p_0_65=[np.nan,0.032,0.046,0.065]
en_2p_0_65=[1.352,0.768,0.529,0.373]
err_en_2p_0_65=[0.077,0.026,0.019,0.028]
```

```python
xdata = np.linspace(0,1/6,20)

plt.figure(num=None, figsize=(8, 6), dpi=160, facecolor='w', edgecolor='k')
plt.errorbar(x_vol,en_gr_0_55,err_en_gr_0_55,fmt='o',label = 'ground state')
plt.errorbar(x_vol,en_ex_0_55,err_en_ex_0_55,fmt='s',label = 'excited state')
plt.errorbar(x_vol,en_2p_0_55,err_en_2p_0_55,fmt='x',label = '2 momentum state')
#plt.yscale("log")
#plt.plot(xdata,E_c(xdata),label='$E_{th}$')
#plt.plot(xdata,E2_c(xdata),label='$E2_{th}$')
plt.plot(xdata,E_l(xdata),label='$E_{p}$')
plt.plot(xdata,E2_l(xdata),label='$E_{2p}$')
plt.plot(xdata,E_m(xdata,0.5),label='$E_{m}$')
plt.xlabel('a/L')
plt.ylabel(r'$aE$')
plt.xlim([0,1/6])
#plt.ylim([-0.02,0.02])
plt.legend()
plt.title("Energy from variational analysis, $\kappa = 0.55$")
plt.savefig("Volume_plot_k_0_55.eps")
```
